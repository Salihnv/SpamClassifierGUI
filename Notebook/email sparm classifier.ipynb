```python
df.drop(df.index[(df['Category']=='{"mode":"full"')],axis=0,inplace=True)
df
```

```python
df['Category']=df['Category'].map({'spam':1,'ham':0})
df
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.corpus import stopwords
lst=stopwords.words('english')
lst
```

```python
mail=df['Message']
mail
```

```python
from nltk import TweetTokenizer
tk=TweetTokenizer()
mail=mail.apply(lambda x:tk.tokenize(x)).apply(lambda x:' '.join(x))
mail
```

```python

def remove_special_characters(text):
    return re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
mail = mail.apply(remove_special_characters)
mail
```

```python
from nltk.tokenize import word_tokenize
mail=mail.apply(lambda x:' '.join([w for w in word_tokenize(x) if len(w)>=3]))
mail
```

```python
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer('english')
mail=mail.apply(lambda x:[stemmer.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda x:' '.join(x))
mail
```

```python
from nltk.corpus import stopwords
sw=stopwords.words('english')
mail=mail.apply(lambda x:[i for i in tk.tokenize(x) if i not in sw]).apply(lambda x:' '.join(x))
mail
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
train_data=vec.fit_transform(mail)
train_data
```

```python
train_data.shape
```

```python
print(train_data)
```

```python
x=train_data
y=df['Category']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)

from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
```

```python
import joblib
joblib.dump(model, 'svm_model.pkl')
```
